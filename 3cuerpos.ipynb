```python
%pylab inline

#Este programa realiza el cálculo por medio del método de Rugen-Kutta para el problema de los tres cuerpos
#Se tienen tres cuerpos que definen un triángulo dentro de una circunferencia y velocidades perpendiculares al plano(condiciones iniciales)
#Definamos las variables a utilizar:

#Masas (Unidad de masa solar)

m1=40.0
m2=20.0
m3=15.0
#Constante de gravitación Universal (unidad de ua+3 msun-1 yr-1)

G=39.486

#Radio inicial (convencion)

R=90.0

#Velocidad incial (convencion)

v=0.5

#Definamos una separación para los intervalos de tiempo

h=0.1

min_t=0.0
max_t=50.0

n_points=int((max_t-min_t)/h)

t=zeros(n_points)

#Haremos la identificación de y1 con las funciones de posición y y2 con las funciones de velocidad de esta manera xi_y1 corresponde a la posición en x de la partícula iésima y análogo para las otras posiciones y y2.
#Posiciones:

x1_y1=zeros(n_points)
y1_y1=zeros(n_points)
z1_y1=zeros(n_points)

x2_y1=zeros(n_points)
y2_y1=zeros(n_points)
z2_y1=zeros(n_points)

x3_y1=zeros(n_points)
y3_y1=zeros(n_points)
z3_y1=zeros(n_points)

#Velocidades:

x1_y2=zeros(n_points)
y1_y2=zeros(n_points)
z1_y2=zeros(n_points)

x2_y2=zeros(n_points)
y2_y2=zeros(n_points)
z2_y2=zeros(n_points)

x3_y2=zeros(n_points)
y3_y2=zeros(n_points)
z3_y2=zeros(n_points)


#Condiciones iniciales:

t[0]=min_t

#Cuerpo 1

x1_y1[0]=0.0
y1_y1[0]=R
z1_y1[0]=0.0

x1_y2[0]=0.0
y1_y2[0]=0.0
z1_y2[0]=v

#Cuerpo 2

x2_y1[0]=0.5*R*(3**0.5)
y2_y1[0]=-0.5*R
z2_y1[0]=0.0

x2_y2[0]=0.0
y2_y2[0]=0.0
z2_y2[0]=v/2

#Cuerpo 3

x3_y1[0]=-0.5*R*(3**0.5)
y3_y1[0]=-0.5*R
z3_y1[0]=0.0

x3_y2[0]=0.0
y3_y2[0]=0.0
z3_y2[0]=0.0

#Resolvemos las 9 ecuaciones diferenciales al mismo tiempo:

for j in range(1,n_points):


	#Ecuacion dif. para x1

	def x1_function_prime_1(t,var_x1_y1,var_x1_y2):
		return var_x1_y2

	def x1_function_prime_2(t,var_x1_y1,var_x1_y2):
		return (G*m2*(x2_y1[j-1]-var_x1_y1)/((((x2_y1[j-1]-var_x1_y1)**2)+((y2_y1[j-1]-y1_y1[j-1])**2)+((z2_y1[j-1]-z1_y1[j-1])**2)))**1.5)+(G*m3*(x3_y1[j-1]-var_x1_y1)/((((x3_y1[j-1]-var_x1_y1)**2)+((y3_y1[j-1]-y1_y1[j-1])**2)+((z3_y1[j-1]-z1_y1[j-1])**2)))**1.5)
	
	#Ecuacion dif. para x2	

	def x2_function_prime_1(t,var_x2_y1,var_x2_y2):
		return var_x2_y2

	def x2_function_prime_2(t,var_x2_y1,var_x2_y2):
		return (-G*m1*(var_x2_y1-x1_y1[j-1])/((((var_x2_y1-x1_y1[j-1])**2)+((y2_y1[j-1]-y1_y1[j-1])**2)+((z2_y1[j-1]-z1_y1[j-1])**2)))**1.5)+(G*m3*(x3_y1[j-1]-var_x2_y1)/((((x3_y1[j-1]-var_x2_y1)**2)+((y3_y1[j-1]-y2_y1[j-1])**2)+((z3_y1[j-1]-z2_y1[j-1])**2)))**1.5)

	#Ecuacion dif. para x3

	def x3_function_prime_1(t,var_x3_y1,var_x3_y2):
		return var_x3_y2

	def x3_function_prime_2(t,var_x3_y1,var_x3_y2):
		return (-G*m2*(var_x3_y1-x2_y1[j-1])/((((var_x3_y1-x2_y1[j-1])**2)+((y3_y1[j-1]-y2_y1[j-1])**2)+((z3_y1[j-1]-z2_y1[j-1])**2)))**1.5)+(-G*m1*(var_x3_y1-x1_y1[j-1])/((((var_x3_y1-x1_y1[j-1])**2)+((y3_y1[j-1]-y1_y1[j-1])**2)+((z3_y1[j-1]-z1_y1[j-1])**2)))**1.5)
	
	
	#Ecuacion dif. para y1

	def y1_function_prime_1(t,var_y1_y1,var_y1_y2):
		return var_y1_y2

	def y1_function_prime_2(t,var_y1_y1,var_y1_y2):
		return (G*m2*(y2_y1[j-1]-var_y1_y1)/((((x2_y1[j-1]-x1_y1[j-1])**2)+((y2_y1[j-1]-var_y1_y1)**2)+((z2_y1[j-1]-z1_y1[j-1])**2)))**1.5)+(G*m3*(y3_y1[j-1]-var_y1_y1)/((((x3_y1[j-1]-x1_y1[j-1])**2)+((y3_y1[j-1]-var_y1_y1)**2)+((z3_y1[j-1]-z1_y1[j-1])**2)))**1.5)
	
	#Ecuacion dif. para y2	

	def y2_function_prime_1(t,var_y2_y1,var_y2_y2):
		return var_y2_y2

	def y2_function_prime_2(t,var_y2_y1,var_y2_y2):
		return (-G*m1*(var_y2_y1-y1_y1[j-1])/((((x2_y1[j-1]-x1_y1[j-1])**2)+((var_y2_y1-y1_y1[j-1])**2)+((z2_y1[j-1]-z1_y1[j-1])**2)))**1.5)+(G*m3*(y3_y1[j-1]-var_y2_y1)/((((x3_y1[j-1]-x2_y1[j-1])**2)+((y3_y1[j-1]-var_y2_y1)**2)+((z3_y1[j-1]-z2_y1[j-1])**2)))**1.5)

	#Ecuacion dif. para y3

	def y3_function_prime_1(t,var_y3_y1,var_y3_y2):
		return var_y3_y2

	def y3_function_prime_2(t,var_y3_y1,var_y3_y2):
		return (-G*m2*(var_y3_y1-y2_y1[j-1])/((((x3_y1[j-1]-x2_y1[j-1])**2)+((var_y3_y1-y2_y1[j-1])**2)+((z3_y1[j-1]-z2_y1[j-1])**2)))**1.5)+(-G*m1*(var_y3_y1-y1_y1[j-1])/((((x3_y1[j-1]-x1_y1[j-1])**2)+((var_y3_y1-y1_y1[j-1])**2)+((z3_y1[j-1]-z1_y1[j-1])**2)))**1.5)
	
	
	#Ecuacion dif. para z1

	def z1_function_prime_1(t,var_z1_y1,var_z1_y2):
		return var_z1_y2

	def z1_function_prime_2(t,var_z1_y1,var_z1_y2):
		return (G*m2*(z2_y1[j-1]-var_z1_y1)/((((x2_y1[j-1]-x1_y1[j-1])**2)+((y2_y1[j-1]-y1_y1[j-1])**2)+((z2_y1[j-1]-var_z1_y1)**2)))**1.5)+(G*m3*(z3_y1[j-1]-var_z1_y1)/((((x3_y1[j-1]-x1_y1[j-1])**2)+((y3_y1[j-1]-y1_y1[j-1])**2)+((z3_y1[j-1]-var_z1_y1)**2)))**1.5)
	
	#Ecuacion dif. para z2	

	def z2_function_prime_1(t,var_z2_y1,var_z2_y2):
		return var_z2_y2

	def z2_function_prime_2(t,var_z2_y1,var_z2_y2):
		return (-G*m1*(var_z2_y1-z1_y1[j-1])/((((x2_y1[j-1]-x1_y1[j-1])**2)+((y2_y1[j-1]-y1_y1[j-1])**2)+((var_z2_y1-z1_y1[j-1])**2)))**1.5)+(G*m3*(z3_y1[j-1]-var_z2_y1)/((((x3_y1[j-1]-x2_y1[j-1])**2)+((y3_y1[j-1]-y2_y1[j-1])**2)+((z3_y1[j-1]-var_z2_y1)**2)))**1.5)

	#Ecuacion dif. para z3

	def z3_function_prime_1(t,var_z3_y1,var_z3_y2):
		return var_z3_y2

	def z3_function_prime_2(t,var_z3_y1,var_z3_y2):
		return (-G*m2*(var_z3_y1-z2_y1[j-1])/((((x3_y1[j-1]-x2_y1[j-1])**2)+((y3_y1[j-1]-y2_y1[j-1])**2)+((var_z3_y1-z2_y1[j-1])**2)))**1.5)+(-G*m1*(var_z3_y1-z1_y1[j-1])/((((x3_y1[j-1]-x1_y1[j-1])**2)+((y3_y1[j-1]-y1_y1[j-1])**2)+((var_z3_y1-z1_y1[j-1])**2)))**1.5)
	

	#Definimos el proceso de Runge-Kutta:

	def RungeKuttaFourthOrderStep(t_old,x1_y1_old,x1_y2_old,x2_y1_old,x2_y2_old,x3_y1_old,x3_y2_old, y1_y1_old,y1_y2_old,y2_y1_old,y2_y2_old,y3_y1_old,y3_y2_old,z1_y1_old,z1_y2_old,z2_y1_old,z2_y2_old,z3_y1_old,z3_y2_old):


		x1_k1_prime1 = x1_function_prime_1(t_old,x1_y1_old,x1_y2_old)
    		x1_k1_prime2 = x1_function_prime_2(t_old,x1_y1_old,x1_y2_old)
		
		x2_k1_prime1 = x2_function_prime_1(t_old,x2_y1_old,x2_y2_old)
    		x2_k1_prime2 = x2_function_prime_2(t_old,x2_y1_old,x2_y2_old)

		x3_k1_prime1 = x3_function_prime_1(t_old,x3_y1_old,x3_y2_old)
    		x3_k1_prime2 = x3_function_prime_2(t_old,x3_y1_old,x3_y2_old)

		y1_k1_prime1 = y1_function_prime_1(t_old,y1_y1_old,y1_y2_old)
    		y1_k1_prime2 = y1_function_prime_2(t_old,y1_y1_old,y1_y2_old)
		
		y2_k1_prime1 = y2_function_prime_1(t_old,y2_y1_old,y2_y2_old)
    		y2_k1_prime2 = y2_function_prime_2(t_old,y2_y1_old,y2_y2_old)

		y3_k1_prime1 = y3_function_prime_1(t_old,y3_y1_old,y3_y2_old)
    		y3_k1_prime2 = y3_function_prime_2(t_old,y3_y1_old,y3_y2_old)

		z1_k1_prime1 = z1_function_prime_1(t_old,z1_y1_old,z1_y2_old)
    		z1_k1_prime2 = z1_function_prime_2(t_old,z1_y1_old,z1_y2_old)
		
		z2_k1_prime1 = z2_function_prime_1(t_old,z2_y1_old,z2_y2_old)
    		z2_k1_prime2 = z2_function_prime_2(t_old,z2_y1_old,z2_y2_old)

		z3_k1_prime1 = z3_function_prime_1(t_old,z3_y1_old,z3_y2_old)
    		z3_k1_prime2 = z3_function_prime_2(t_old,z3_y1_old,z3_y2_old)

		#Primer paso

    		t1 = t_old+ (h/2.0)

    		x1_y1_1 = x1_y1_old + ((h/2.0) * x1_k1_prime1)
   		x1_y2_1 = x1_y2_old + ((h/2.0) * x1_k1_prime2)

		x2_y1_1 = x2_y1_old + ((h/2.0) * x2_k1_prime1)
   		x2_y2_1 = x2_y2_old + ((h/2.0) * x2_k1_prime2)

		x3_y1_1 = x3_y1_old + ((h/2.0) * x3_k1_prime1)
   		x3_y2_1 = x3_y2_old + ((h/2.0) * x3_k1_prime2)

		y1_y1_1 = y1_y1_old + ((h/2.0) * y1_k1_prime1)
   		y1_y2_1 = y1_y2_old + ((h/2.0) * y1_k1_prime2)

		y2_y1_1 = y2_y1_old + ((h/2.0) * y2_k1_prime1)
   		y2_y2_1 = y2_y2_old + ((h/2.0) * y2_k1_prime2)

		y3_y1_1 = y3_y1_old + ((h/2.0) * y3_k1_prime1)
   		y3_y2_1 = y3_y2_old + ((h/2.0) * y3_k1_prime2)

		z1_y1_1 = z1_y1_old + ((h/2.0) * z1_k1_prime1)
   		z1_y2_1 = z1_y2_old + ((h/2.0) * z1_k1_prime2)

		z2_y1_1 = z2_y1_old + ((h/2.0) * z2_k1_prime1)
   		z2_y2_1 = z2_y2_old + ((h/2.0) * z2_k1_prime2)

		z3_y1_1 = z3_y1_old + ((h/2.0) * z3_k1_prime1)
   		z3_y2_1 = z3_y2_old + ((h/2.0) * z3_k1_prime2)

   
		x1_k2_prime1 = x1_function_prime_1(t1, x1_y1_1, x1_y2_1)
    		x1_k2_prime2 = x1_function_prime_2(t1, x1_y1_1, x1_y2_1)

		x2_k2_prime1 = x2_function_prime_1(t1, x2_y1_1, x2_y2_1)
    		x2_k2_prime2 = x2_function_prime_2(t1, x2_y1_1, x2_y2_1)

		x3_k2_prime1 = x3_function_prime_1(t1, x3_y1_1, x3_y2_1)
    		x3_k2_prime2 = x3_function_prime_2(t1, x3_y1_1, x3_y2_1)

		y1_k2_prime1 = y1_function_prime_1(t1, y1_y1_1, y1_y2_1)
    		y1_k2_prime2 = y1_function_prime_2(t1, y1_y1_1, y1_y2_1)

		y2_k2_prime1 = y2_function_prime_1(t1, y2_y1_1, y2_y2_1)
    		y2_k2_prime2 = y2_function_prime_2(t1, y2_y1_1, y2_y2_1)

		y3_k2_prime1 = y3_function_prime_1(t1, y3_y1_1, y3_y2_1)
    		y3_k2_prime2 = y3_function_prime_2(t1, y3_y1_1, y3_y2_1)

		z1_k2_prime1 = z1_function_prime_1(t1, z1_y1_1, z1_y2_1)
    		z1_k2_prime2 = z1_function_prime_2(t1, z1_y1_1, z1_y2_1)

		z2_k2_prime1 = z2_function_prime_1(t1, z2_y1_1, z2_y2_1)
    		z2_k2_prime2 = z2_function_prime_2(t1, z2_y1_1, z2_y2_1)

		z3_k2_prime1 = z3_function_prime_1(t1, z3_y1_1, z3_y2_1)
    		z3_k2_prime2 = z3_function_prime_2(t1, z3_y1_1, z3_y2_1)
    
    		#Segundo paso

   		t2 = t_old + (h/2.0)

    		x1_y1_2 = x1_y1_old + ((h/2.0) * (x1_k2_prime1))
    		x1_y2_2 = x1_y2_old + ((h/2.0) * (x1_k2_prime2))

		x2_y1_2 = x2_y1_old + ((h/2.0) * (x2_k2_prime1))
    		x2_y2_2 = x2_y2_old + ((h/2.0) * (x2_k2_prime2))

		x3_y1_2 = x3_y1_old + ((h/2.0) * (x3_k2_prime1))
    		x3_y2_2 = x3_y2_old + ((h/2.0) * (x3_k2_prime2))

		y1_y1_2 = y1_y1_old + ((h/2.0) * (y1_k2_prime1))
    		y1_y2_2 = y1_y2_old + ((h/2.0) * (y1_k2_prime2))

		y2_y1_2 = y2_y1_old + ((h/2.0) * (y2_k2_prime1))
    		y2_y2_2 = y2_y2_old + ((h/2.0) * (y2_k2_prime2))

		y3_y1_2 = y3_y1_old + ((h/2.0) * (y3_k2_prime1))
    		y3_y2_2 = y3_y2_old + ((h/2.0) * (y3_k2_prime2))

		z1_y1_2 = z1_y1_old + ((h/2.0) * (z1_k2_prime1))
    		z1_y2_2 = z1_y2_old + ((h/2.0) * (z1_k2_prime2))

		z2_y1_2 = z2_y1_old + ((h/2.0) * (z2_k2_prime1))
    		z2_y2_2 = z2_y2_old + ((h/2.0) * (z2_k2_prime2))

		z3_y1_2 = z3_y1_old + ((h/2.0) * (z3_k2_prime1))
    		z3_y2_2 = z3_y2_old + ((h/2.0) * (z3_k2_prime2))


		x1_k3_prime1 = x1_function_prime_1(t2, x1_y1_2, x1_y2_2)
   		x1_k3_prime2 = x1_function_prime_2(t2, x1_y1_2, x1_y2_2)
    		
		x2_k3_prime1 = x2_function_prime_1(t2, x2_y1_2, x2_y2_2)
   		x2_k3_prime2 = x2_function_prime_2(t2, x2_y1_2, x2_y2_2)
    
		x3_k3_prime1 = x3_function_prime_1(t2, x3_y1_2, x3_y2_2)
   		x3_k3_prime2 = x3_function_prime_2(t2, x3_y1_2, x3_y2_2)

		y1_k3_prime1 = y1_function_prime_1(t2, y1_y1_2, y1_y2_2)
   		y1_k3_prime2 = y1_function_prime_2(t2, y1_y1_2, y1_y2_2)
    		
		y2_k3_prime1 = y2_function_prime_1(t2, y2_y1_2, y2_y2_2)
   		y2_k3_prime2 = y2_function_prime_2(t2, y2_y1_2, y2_y2_2)
    
		y3_k3_prime1 = y3_function_prime_1(t2, y3_y1_2, y3_y2_2)
   		y3_k3_prime2 = y3_function_prime_2(t2, y3_y1_2, y3_y2_2)

		z1_k3_prime1 = z1_function_prime_1(t2, z1_y1_2, z1_y2_2)
   		z1_k3_prime2 = z1_function_prime_2(t2, z1_y1_2, z1_y2_2)
    		
		z2_k3_prime1 = z2_function_prime_1(t2, z2_y1_2, z2_y2_2)
   		z2_k3_prime2 = z2_function_prime_2(t2, z2_y1_2, z2_y2_2)
    
		z3_k3_prime1 = z3_function_prime_1(t2, z3_y1_2, z3_y2_2)
   		z3_k3_prime2 = z3_function_prime_2(t2, z3_y1_2, z3_y2_2)
    
   		#Tercer paso

    		t3 = t_old + h

    		x1_y1_3 = x1_y1_old + (h * (x1_k3_prime1))
   		x1_y2_3 = x1_y2_old + (h * (x1_k3_prime2))

		x2_y1_3 = x2_y1_old + (h * (x2_k3_prime1))
   		x2_y2_3 = x2_y2_old + (h * (x2_k3_prime2))

		x3_y1_3 = x3_y1_old + (h * (x3_k3_prime1))
   		x3_y2_3 = x3_y2_old + (h * (x3_k3_prime2))

		y1_y1_3 = y1_y1_old + (h * (y1_k3_prime1))
   		y1_y2_3 = y1_y2_old + (h * (y1_k3_prime2))

		y2_y1_3 = y2_y1_old + (h * (y2_k3_prime1))
   		y2_y2_3 = y2_y2_old + (h * (y2_k3_prime2))

		y3_y1_3 = y3_y1_old + (h * (y3_k3_prime1))
   		y3_y2_3 = y3_y2_old + (h * (y3_k3_prime2))

		z1_y1_3 = z1_y1_old + (h * (z1_k3_prime1))
   		z1_y2_3 = z1_y2_old + (h * (z1_k3_prime2))

		z2_y1_3 = z2_y1_old + (h * (z2_k3_prime1))
   		z2_y2_3 = z2_y2_old + (h * (z2_k3_prime2))

		z3_y1_3 = z3_y1_old + (h * (z3_k3_prime1))
   		z3_y2_3 = z3_y2_old + (h * (z3_k3_prime2))


    		x1_k4_prime1 = x1_function_prime_1(t3, x1_y1_3, x1_y2_3)
    		x1_k4_prime2 = x1_function_prime_2(t3, x1_y1_3, x1_y2_3)

		x2_k4_prime1 = x2_function_prime_1(t3, x2_y1_3, x2_y2_3)
    		x2_k4_prime2 = x2_function_prime_2(t3, x2_y1_3, x2_y2_3)

		x3_k4_prime1 = x3_function_prime_1(t3, x3_y1_3, x3_y2_3)
    		x3_k4_prime2 = x3_function_prime_2(t3, x3_y1_3, x3_y2_3)

		y1_k4_prime1 = y1_function_prime_1(t3, y1_y1_3, y1_y2_3)
    		y1_k4_prime2 = y1_function_prime_2(t3, y1_y1_3, y1_y2_3)

		y2_k4_prime1 = y2_function_prime_1(t3, y2_y1_3, y2_y2_3)
    		y2_k4_prime2 = y2_function_prime_2(t3, y2_y1_3, y2_y2_3)

		y3_k4_prime1 = y3_function_prime_1(t3, y3_y1_3, y3_y2_3)
    		y3_k4_prime2 = y3_function_prime_2(t3, y3_y1_3, y3_y2_3)

		z1_k4_prime1 = z1_function_prime_1(t3, z1_y1_3, z1_y2_3)
    		z1_k4_prime2 = z1_function_prime_2(t3, z1_y1_3, z1_y2_3)

		z2_k4_prime1 = z2_function_prime_1(t3, z2_y1_3, z2_y2_3)
    		z2_k4_prime2 = z2_function_prime_2(t3, z2_y1_3, z2_y2_3)

		z3_k4_prime1 = z3_function_prime_1(t3, z3_y1_3, z3_y2_3)
    		z3_k4_prime2 = z3_function_prime_2(t3, z3_y1_3, z3_y2_3)
    
    		#Cuarto paso

   		x1_average_k1 = (1.0/6.0)*(x1_k1_prime1 + 2.0*x1_k2_prime1 + 2.0*x1_k3_prime1 + x1_k4_prime1)
   	 	x1_average_k2 = (1.0/6.0)*(x1_k1_prime2 + 2.0*x1_k2_prime2 + 2.0*x1_k3_prime2 + x1_k4_prime2)

		x2_average_k1 = (1.0/6.0)*(x2_k1_prime1 + 2.0*x2_k2_prime1 + 2.0*x2_k3_prime1 + x2_k4_prime1)
   	 	x2_average_k2 = (1.0/6.0)*(x2_k1_prime2 + 2.0*x2_k2_prime2 + 2.0*x2_k3_prime2 + x2_k4_prime2)

		x3_average_k1 = (1.0/6.0)*(x3_k1_prime1 + 2.0*x3_k2_prime1 + 2.0*x3_k3_prime1 + x3_k4_prime1)
   	 	x3_average_k2 = (1.0/6.0)*(x3_k1_prime2 + 2.0*x3_k2_prime2 + 2.0*x3_k3_prime2 + x3_k4_prime2)

		y1_average_k1 = (1.0/6.0)*(y1_k1_prime1 + 2.0*y1_k2_prime1 + 2.0*y1_k3_prime1 + y1_k4_prime1)
   	 	y1_average_k2 = (1.0/6.0)*(y1_k1_prime2 + 2.0*y1_k2_prime2 + 2.0*y1_k3_prime2 + y1_k4_prime2)

		y2_average_k1 = (1.0/6.0)*(y2_k1_prime1 + 2.0*y2_k2_prime1 + 2.0*y2_k3_prime1 + y2_k4_prime1)
   	 	y2_average_k2 = (1.0/6.0)*(y2_k1_prime2 + 2.0*y2_k2_prime2 + 2.0*y2_k3_prime2 + y2_k4_prime2)

		y3_average_k1 = (1.0/6.0)*(y3_k1_prime1 + 2.0*y3_k2_prime1 + 2.0*y3_k3_prime1 + y3_k4_prime1)
   	 	y3_average_k2 = (1.0/6.0)*(y3_k1_prime2 + 2.0*y3_k2_prime2 + 2.0*y3_k3_prime2 + y3_k4_prime2)

		z1_average_k1 = (1.0/6.0)*(z1_k1_prime1 + 2.0*z1_k2_prime1 + 2.0*z1_k3_prime1 + z1_k4_prime1)
   	 	z1_average_k2 = (1.0/6.0)*(z1_k1_prime2 + 2.0*z1_k2_prime2 + 2.0*z1_k3_prime2 + z1_k4_prime2)

		z2_average_k1 = (1.0/6.0)*(z2_k1_prime1 + 2.0*z2_k2_prime1 + 2.0*z2_k3_prime1 + z2_k4_prime1)
   	 	z2_average_k2 = (1.0/6.0)*(z2_k1_prime2 + 2.0*z2_k2_prime2 + 2.0*z2_k3_prime2 + z2_k4_prime2)

		z3_average_k1 = (1.0/6.0)*(z3_k1_prime1 + 2.0*z3_k2_prime1 + 2.0*z3_k3_prime1 + z3_k4_prime1)
   	 	z3_average_k2 = (1.0/6.0)*(z3_k1_prime2 + 2.0*z3_k2_prime2 + 2.0*z3_k3_prime2 + z3_k4_prime2)
	
		#Resultados:
    
    		t_new = t_old + h

   		x1_y1_new = x1_y1_old + (h * x1_average_k1)
   		x1_y2_new = x1_y2_old + (h * x1_average_k2)

		x2_y1_new = x2_y1_old + (h * x2_average_k1)
   		x2_y2_new = x2_y2_old + (h * x2_average_k2)

		x3_y1_new = x3_y1_old + (h * x3_average_k1)
   		x3_y2_new = x3_y2_old + (h * x3_average_k2)

		y1_y1_new = y1_y1_old + (h * y1_average_k1)
   		y1_y2_new = y1_y2_old + (h * y1_average_k2)

		y2_y1_new = y2_y1_old + (h * y2_average_k1)
   		y2_y2_new = y2_y2_old + (h * y2_average_k2)

		y3_y1_new = y3_y1_old + (h * y3_average_k1)
   		y3_y2_new = y3_y2_old + (h * y3_average_k2)

		z1_y1_new = z1_y1_old + (h * z1_average_k1)
   		z1_y2_new = z1_y2_old + (h * z1_average_k2)

		z2_y1_new = z2_y1_old + (h * z2_average_k1)
   		z2_y2_new = z2_y2_old + (h * z2_average_k2)

		z3_y1_new = z3_y1_old + (h * z3_average_k1)
   		z3_y2_new = z3_y2_old + (h * z3_average_k2)

		#Lo que retorna el método de Runge-Kutta

    		return t_new, x1_y1_new, x1_y2_new, x2_y1_new, x2_y2_new, x3_y1_new, x3_y2_new, y1_y1_new, y1_y2_new, y2_y1_new, y2_y2_new, y3_y1_new, y3_y2_new, z1_y1_new, z1_y2_new, z2_y1_new, z2_y2_new, z3_y1_new, z3_y2_new
	 
	#Con cada iteración vamos redefiniendo el tiempo y las posiciones y las velocidades para ir modificando las condiciones iniciales de cada proceso

    	t[j],x1_y1[j],x1_y2[j],x2_y1[j],x2_y2[j],x3_y1[j],x3_y2[j],y1_y1[j],y1_y2[j],y2_y1[j],y2_y2[j],y3_y1[j],y3_y2[j],z1_y1[j],z1_y2[j],z2_y1[j],z2_y2[j],z3_y1[j],z3_y2[j] = RungeKuttaFourthOrderStep(t[j-1],x1_y1[j-1],x1_y2[j-1],x2_y1[j-1],x2_y2[j-1],x3_y1[j-1],x3_y2[j-1],y1_y1[j-1],y1_y2[j-1],y2_y1[j-1],y2_y2[j-1],y3_y1[j-1],y3_y2[j-1],z1_y1[j-1],z1_y2[j-1],z2_y1[j-1],z2_y2[j-1],z3_y1[j-1],z3_y2[j-1])

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x1_y1,y1_y1,z1_y1,c='r')
ax.plot(x2_y1,y2_y1,z2_y1,c='g')
ax.plot(x3_y1,y3_y1,z3_y1,c='b')
```
